### Binary tree data structure
#### It has usually O(logN) in insertion,deletion and searching using key. Worst case O(N)
#### O(N) has a huge performance impact on searching, thus want to be avoided if possible


```python
class Node():
    def __init__(self,content):
        self.content = content
        self.parent  = False
        self.right   = False
        self.left    = False
    def isLeafNode(self):
        return self.right == False and self.left == False
    
    def display_node(self):
        print(self.content)
```

```python
a = Node(3)
a.display_node()
b = Node(4)
a.right = Node(4)
b.display_node()
a.right.display_node()
print(a.isLeafNode())
print(b.isLeafNode())
```

## Insertion of binary tree

```python
import queue
class BST:
    def __init__(self):
        self.root = None
        
    def insert(self,node):
        # If the BST is empty, assign as root node.
        if self.root is None:
            self.root = node
        
        currentNode = self.root
        # Keep Doing the following Until reaching the leaf node.
        while currentNode is not False:
            if node.content > currentNode.content:
                if currentNode.right == False:
                    node.parent = currentNode
                    currentNode.right = node
                    break
                
                # If it is larger than the currentNode value, traverse the right subtree
                currentNode = currentNode.right
            elif node.content < currentNode.content:
                if currentNode.left == False:
                    node.parent = currentNode
                    currentNode.left = node
                    break
                
                # If it is smaller than the currentNode value, traverse the left subtree
                currentNode = currentNode.left
            else:
                # If it is equal to the currentNode value, tell the user, it already exists
                print("Node already exists")
                return False
        
    
    def level_order_traversal(self):
        print("level order traversal")
        if self.root is None:
            return False

        # Create an empty queue for traversing the tree
        q = queue.Queue()
        q.put(self.root)
                
        while not q.empty():
            # Get the front element from the queue
            node = q.get()
            print(node.content, end="\n")

            # Enqueue the left child if it exists
            if node.left is not False:
                q.put(node.left)

            # Enqueue the right child if it exists
            if node.right is not False:
                q.put(node.right)
    
    def pre_order_driver(self):
        print("Pre-order traversal")
        self.pre_order_traversal(self.root)
    
    def pre_order_traversal(self,node):
        if node is False:
            return

        print(node.content, end="\n")
        
        self.pre_order_traversal(node.left)
        self.pre_order_traversal(node.right)
            
```

### Search for node

```python
import queue
class BST:
    def __init__(self):
        self.root = None
        
    def insert(self,node):
        # If the BST is empty, assign as root node.
        if self.root is None:
            self.root = node
        
        currentNode = self.root
        # Keep Doing the following Until reaching the leaf node.
        while currentNode is not False:
            if node.content > currentNode.content:
                if currentNode.right == False:
                    # print("Node Does not exists, Inserting Node")
                    currentNode.right = node
                    node.parent       = currentNode   
                    break
                
                # If it is larger than the currentNode value, traverse the right subtree
                currentNode = currentNode.right
            elif node.content < currentNode.content:
                if currentNode.left == False:
                    # print("Node Does not exists, Inserting Node")
                    currentNode.left = node
                    node.parent       = currentNode   
                    break
                
                # If it is smaller than the currentNode value, traverse the left subtree
                currentNode = currentNode.left
            else:
                # If it is equal to the currentNode value, tell the user, it already exists
                print("Node already exists")
                return False
    
    def level_order_traversal(self):
        print("level order traversal")
        if self.root is None:
            return False

        # Create an empty queue for traversing the tree
        q = queue.Queue()
        q.put(self.root)
                
        while not q.empty():
            # Get the front element from the queue
            node = q.get()
            print(node.content, end="\n")

            # Enqueue the left child if it exists
            if node.left is not False:
                q.put(node.left)

            # Enqueue the right child if it exists
            if node.right is not False:
                q.put(node.right)
                
    def remove_node(self,node):
        # Check if the tree is empty
        if self.root is None:
            return
        
        
        currentNode = self.root
        
        # Keep seraching until you met empty node
        while currentNode is not False:
           if node.content > currentNode.content:
               currentNode = currentNode.right
           elif node.content < currentNode.content:
               currentNode = currentNode.left
           else:
                # Node found, start the removal, determine 
                print("Node Found start deletion!", end="\n\n")
                if currentNode.right is not False and currentNode.left is not False:
                    node_to_swap = currentNode.left
                    # Find the Maximum of left subtree by keep searching the rightist node, until meeting False
                    print("Searching Rightist of left subtree", end="\n\n")
                    while node_to_swap.right is not False:
                        node_to_swap =  node_to_swap.right 
                    
                    # Swap with the current Node then remove this node, just simply swap the content~
                    currentNode.content = node_to_swap.content
                    currentNode.left    = node_to_swap.left
                    
                    
                    return
                elif currentNode.left is not False:
                    print("Swap with left node", end="\n\n")
                    node_to_swap = currentNode.left
                    # Swap with the left node
                    currentNode.right   = node_to_swap.right
                    currentNode.left    = node_to_swap.left
                    currentNode.content = node_to_swap.content
                    
                    return
                elif currentNode.right is not False:
                    print("Swap with right node", end="\n\n")
                    node_to_swap = currentNode.right
                    # Swap with the right node
                    currentNode.right   = node_to_swap.right
                    currentNode.left    = node_to_swap.left
                    currentNode.content = node_to_swap.content
                                
                    return
                else:
                    print("Destroying current Node", end="\n\n")
                    parentNode = currentNode.parent
                    if currentNode.content < parentNode.content:
                        parentNode.left = False
                    else:
                        parentNode.right = False
                    
                    return
        
        print("Node does not exist")
        return 
       
                         
    def pre_order_driver(self):
        print("Pre-order traversal")
        self.pre_order_traversal(self.root)
    
    def pre_order_traversal(self,node):
        if node is False:
            return

        print(node.content, end="\n")
        
        self.pre_order_traversal(node.left)
        self.pre_order_traversal(node.right)
            
```
